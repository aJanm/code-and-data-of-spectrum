```python
# https://blog.csdn.net/Joseph__Lagrange/article/details/95298415
# 区间偏最小二乘法python实现
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# from utility_spectrum import spxy, ipls
import scipy.io as sio
from sklearn.cross_decomposition import PLSRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from math import log
from matplotlib.pyplot import MultipleLocator
from matplotlib import cm  # 输出颜色设置模块
import csv
plt.rcParams['font.sans-serif'] = ['SimHei']   # 用黑体显示中文
plt.rcParams['axes.unicode_minus'] = False     # 正常显示负号
# 获得mse后，绘制柱状图,定义渐变色
# https://www.freesion.com/article/27781348355/
# https://www.freesion.com/article/27781348355/
x_major_locator = MultipleLocator(1)
ax = plt.gca()  # get current axes
ax.xaxis.set_major_locator(x_major_locator)

# 划分不同的光谱区间，找到最适合训练的区间


def splitspectrum(interval_num, x_train, x_test):
    """
    :param interval_num:  int (common values are 10, 20, 30 or 40)
    :param x_train:  shape (n_samples, n_features)
    :param x_test:  shape (n_samples, n_features)
    :return: x_train_block:intervals splitting for training sets（dict）
            x_test_black： intervals splitting for test sets （dict）
    """
    # 尝试将其转为array类型
    x_train, x_test = np.array(x_train.values), np.array(x_test.values)
    feature_num = x_train.shape[1]

    x_train_block = {}
    x_test_black = {}
    remaining = feature_num % interval_num  # 用于检查是否能等分
    # （一）特征数量能够等分的情况
    if not remaining:
        interval_size = feature_num / interval_num  # 子区间波点数量
        for i in range(1, interval_num+1):
            # （1）取对应子区间的光谱数据
            feature_start, feature_end = int(
                (i-1) * interval_size), int(i * interval_size)
            x_train_block[str(i)] = x_train[:, feature_start:feature_end]
            x_test_black[str(i)] = x_test[:, feature_start:feature_end]

    # （二）特征数量不能等分的情况(将多余波点等分到后面的几个区间里)
    else:
        separation = interval_num - remaining  # 前几个区间
        intervalsize1 = feature_num // interval_num
        intervalsize2 = feature_num // interval_num + 1

        # （2）前几个子区间(以separation为界)
        for i in range(1, separation+1):
            feature_start, feature_end = int(
                (i-1) * intervalsize1), int(i * intervalsize1)
            x_train_block[str(i)] = x_train[:, feature_start:feature_end]
            x_test_black[str(i)] = x_test[:, feature_start:feature_end]

        # （3）后几个子区间(以separation为界)
        for i in range(separation+1, interval_num+1):
            feature_s = int((i - separation-1) * intervalsize2) + feature_end
            feature_e = int((i - separation) * intervalsize2) + feature_end
            x_train_block[str(i)] = x_train[:, feature_s:feature_e]
            x_test_black[str(i)] = x_test[:, feature_s:feature_e]

    return x_train_block, x_test_black

# 将 splitspectrum方法得到的区间传入到ipls中间进行训练


def ipls(intervals, x_train, x_test, y_train, y_test):
    """
    :param intervals: 区间数量
    :param x_train: shape (n_samples, n_features)
    :param x_test: shape (n_samples, n_features)
    :param y_train: shape (n_samples, )
    :param y_test: shape (n_samples, )
    :return:
    """
    # 注意把光谱划分为完整的空间，否则可能出错，如果无法划分的话，可以把部分数据列删除
    x_train_block, x_test_black = splitspectrum(intervals, x_train, x_test)

    mse = []
    mae = []
    r2score_y_test = []  # 计算y_test和y_test_pred对应回归的决定系数R2
    r2score_y_train = []
    for i in range(1, intervals + 1):
        print("当前区间:", i)
        x_train_interval, x_test_interval = x_train_block[str(
            i)], x_test_black[str(i)]

        current_fn = x_train_interval.shape[1]
        if current_fn >= 100:
            ncom_upper = 100
        elif current_fn >= 50:
            ncom_upper = current_fn - 10
        else:
            ncom_upper = current_fn - 5
        ncomp = np.arange(5, ncom_upper)

        error = []
        # 2023年1月27日新增rmse和r2score评价参数
        mae_error = []
        r2score_error = []

        # 2023年2月14日
        r2score_y_train_error = []
        # 求出不同主成分对应的误差，并将最小误差添加到mse中
        for nc in ncomp:
            print("迭代当前主成分数量:", nc)
            pls = PLSRegression(n_components=nc,
                                scale=True,
                                max_iter=500,
                                tol=1e-06,
                                copy=True)
            pls.fit(x_train_interval, y_train.values.reshape(-1, 1))
            # 求y_test_pred便于求测试集的R2score
            y_test_pred = pls.predict(x_test_interval)
            # 2023年2月14日
            # 求y_train_pred便于求测试集的R2score
            y_train_pred = pls.predict(x_train_interval)
            
            mse_temp = mean_squared_error(y_test, y_test_pred.ravel())
            # 2023年1月27日
            mae_temp = mean_absolute_error(y_test, y_test_pred.ravel())
            print("y_test_pred.shape:", y_test_pred.shape)
            r2score_temp = r2_score(y_test, y_test_pred.ravel())
            # 2023年2月14日
            r2score_y_train_temp = r2_score(y_train, y_train_pred.ravel())

            error.append(mse_temp)
            # 2023年1月27日
            mae_error.append(mae_temp)
            r2score_error.append(r2score_temp)
            # 2023年2月14日
            r2score_y_train_error.append(r2score_y_train_temp)
        mse.append(np.min(error))
        # 2023年1月30日
        mae.append(np.min(mae_error))
        r2score_y_test.append(np.min(r2score_error))
        # 2023年2月14日
        r2score_y_train.append(np.min(r2score_y_train_error))
    # 2023年1月30日在函数外面的用到了mse_log，所以里面可以不需要
    mse_log = []
    for i in range(len(mse)):
        mse_log.append(log(mse[i], 10))
# 对应三种方差
# mse: mean_squared_error
# rmse: mean_absolute_error
# r2score: r2_score
# 数据维度：1行X光谱分段数（比如十段）
    return mse, mae, r2score_y_test, r2score_y_train


# 1.数据获取
'''
mat = sio.loadmat('ndfndf.mat')
data = mat['ndfndf']
x, y = data[:, :1050], data[:, 1050]
'''
data = pd.read_csv('./larger_wavelength_xy.csv')
y = data.iloc[:, 0]
x = data.iloc[:, 1:]
print(x.shape, y.shape)


intervals = 10
# 在这里控制运行的次数
for i in range(50):
    # 2.样本集划分,因为每次划分的测试集训练集不一样所以训练结果可能不一样
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
    print(x_train.shape, y_train.shape)
    print(x_test.shape, y_test.shape)
    mse, mae, r2score_y_test, r2_score_y_train = \
    ipls(intervals, x_train, x_test, y_train, y_test)
    
    mse_log, mae_log, r2score_log = [], [], []
    for i in range(len(mse)):
        mse_log.append(log(mse[i], 10))
        # 归一化
    # 2023年1月30日
    for i in range(len(mae)):
        mae_log.append(log(mae[i], 10))
    # 先不对r2score取对数
    # 2023年1月30日试试不取对数，因为r2score的值应该较小
    # for i in range(len(r2score)):
    #     r2score_log.append(log(r2score[i], 10))
    # mse_log
    # 暂时注释，写入文件的命令
    with open('./mse_data_2023_0214.csv','a',newline='') as f1:
        writer = csv.writer(f1)
        # 每次写入一行10个mse_log数据，
        # 之所以使用log是因为画出来的图效果更明显
        writer.writerow(mse_log)
        f1.close()
    # 2023年1月30日
    with open('./mae_data_2023_0214.csv','a',newline='') as f2:
        writer = csv.writer(f2)
        # 每次写入一行10个mse_log数据，
        # 之所以使用log是因为画出来的图效果更明显
        writer.writerow(mae_log)
        f2.close()
    
    with open('./r2score_test_data_2023_0214.csv','a',newline='') as f3:
        writer = csv.writer(f3)
        # 每次写入一行10个mse_log数据，
        # 之所以使用log是因为画出来的图效果更明显
        # writer.writerow(r2score_log)
        # 2023年1月30日试试不取对数，因为r2score的值应该较小
        writer.writerow(r2score_y_test)
        f3.close()

    with open('./r2score_train_data_2023_0214.csv','a',newline='') as f4:
        writer = csv.writer(f4)
        # 每次写入一行10个mse_log数据，
        # 之所以使用log是因为画出来的图效果更明显
        # writer.writerow(r2score_log)
        # 2023年1月30日试试不取对数，因为r2score的值应该较小
        writer.writerow(r2_score_y_train)
        f4.close()

```

```python
def ipls(intervals, x_train, x_test, y_train, y_test):
    """
    :param intervals: 区间数量
    :param x_train: shape (n_samples, n_features)
    :param x_test: shape (n_samples, n_features)
    :param y_train: shape (n_samples, )
    :param y_test: shape (n_samples, )
    :return:
    """
    x_train_block, x_test_black = splitspectrum(intervals, x_train, x_test)

    mse = []
    for i in range(1, intervals + 1):
        print("当前区间:", i)
        x_train_interval, x_test_interval = x_train_block[str(
            i)], x_test_black[str(i)]

        current_fn = x_train_interval.shape[1]
        if current_fn >= 100:
            ncom_upper = 100
        elif current_fn >= 50:
            ncom_upper = current_fn - 10
        else:
            ncom_upper = current_fn - 5
        ncomp = np.arange(5, ncom_upper)

        error = []
        # 求出不同主成分对应的误差，并将最小误差添加到mse中
        for nc in ncomp:
            print("迭代当前主成分数量:", nc)
            pls = PLSRegression(n_components=nc,
                                scale=True,
                                max_iter=500,
                                tol=1e-06,
                                copy=True)
            pls.fit(x_train_interval, y_train.values.reshape(-1, 1))
            y_test_pred = pls.predict(x_test_interval)
            mse_temp = mean_squared_error(y_test, y_test_pred.ravel())
            error.append(mse_temp)
        mse.append(np.min(error))

    mse_log = []
    for i in range(len(mse)):
        mse_log.append(log(mse[i], 10))

    return mse
```
